```python
df.info()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
cat=df.select_dtypes('object')
num=df.select_dtypes(['int','float'])
cat.head()
```

```python
cat["Workclass"]=le.fit_transform(cat["Workclass"])
cat["education"]=le.fit_transform(cat["education"])
cat["marital-status"]=le.fit_transform(cat["marital-status"])
cat["occupation"]=le.fit_transform(cat["occupation"])
cat["relationship"]=le.fit_transform(cat["relationship"])
cat["native-country"]=le.fit_transform(cat["native-country"])
cat["race"]=le.fit_transform(cat["race"])
cat["sex"]=le.fit_transform(cat["sex"])
```

```python
df=pd.concat([num,cat],axis=1)
```

```python
df.head()
```

```python
x=df.drop('Income',axis=1)
x
y=df['Income']
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df=sc.fit_transform(x)
x=pd.DataFrame(df,columns=x.columns)
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(f'Accuracy-:{accuracy_score(ytest,ypred)}')
print(f'{confusion_matrix(ytest,ypred)}')
print(classification_report(ytest,ypred))
```

```python
#Increasing accuracy
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(solver='liblinear')
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(f'Accuracy-:{accuracy_score(ytest,ypred)}')
print(f'{confusion_matrix(ytest,ypred)}')
print(classification_report(ytest,ypred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
knn=KNeighborsRegressor()
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
```

```python
from sklearn.metrics import classification_report,r2_score,mean_absolute_error,mean_squared_error
print(f'Accuracy-:{r2_score(ytest,ypred)}')
print(f'mean absolute error-:{mean_absolute_error(ytest,ypred)}')
print(f'mean_squared_error-:{mean_squared_error(ytest,ypred)}')
```
